#converting the seniorCitizen column to object as it has value 0 and 1

```python
def conv(value):
    if value == 1:
        return "yes"
    else:
        return "no"
    
df["SeniorCitizen"] = df["SeniorCitizen"].apply(conv)
```

```python
ax = sns.countplot(x = 'Churn', data = df)

ax.bar_label(ax.containers[0])
plt.title("Count of Customers by Churn", fontsize = 10)
plt.show()
```

```python
plt.figure(figsize = (3,4))
gb = df.groupby('Churn').agg({'Churn':"count"})
plt.pie(gb['Churn'], labels = gb.index, autopct = "%1.2f%%")
plt.title("Percentage of Churned Customers", fontsize = 10)
plt.show()
```

#From the given pie chart we can conclude that 26.54% of our customers have churned out
#Now let's explore the reason behind it

```python
plt.figure(figsize = (3, 3))
sns.countplot(x = "gender", data = df, hue = "Churn")
plt.title("Churn by Gender")
plt.show()
```

```python
plt.figure(figsize = (3, 4))
ax = sns.countplot(x = "SeniorCitizen", data = df)
plt.title("Count of Customer as Senior Citizen")
ax.bar_label(ax.containers[0])
plt.show()
```

```python
# Calculate the total count of each category
total = df.groupby('SeniorCitizen')['Churn'].value_counts(normalize=True).unstack() * 100

# Create the stacked bar chart
fig, ax = plt.subplots(figsize=(5, 5))

# Plot the bars for each Churn category
total.plot(kind='bar', stacked=True, ax=ax)

# Add percentage labels to the bars
for p in ax.patches:
    width = p.get_width()  # bar width
    height = p.get_height()  # bar height
    x, y = p.get_xy()  # bar bottom left coordinates
    ax.text(x + width / 2, 
            y + height / 2, 
            f'{height:.1f}%', 
            ha='center', 
            va='center')

# Customize the chart
plt.title("Churn by Senior Citizen (Stacked %)")
plt.ylabel("Percentage (%)")
plt.xticks(rotation=0)
plt.show()
```

```python
# List of columns you want to create count plots for
columns = ['PhoneService', 'MultipleLines', 'InternetService', 
           'OnlineSecurity', 'OnlineBackup', 'DeviceProtection', 
           'TechSupport', 'StreamingTV', 'StreamingMovies']

# Define the number of subplots: 3 rows and 3 columns (since we have 9 features)
fig, axes = plt.subplots(3, 3, figsize=(15, 12))  # 3x3 grid of plots

# Flatten axes array to easily loop through
axes = axes.flatten()

# Create a count plot for each column
for i, col in enumerate(columns):
    sns.countplot(x=col, data=df, ax=axes[i], hue = df['Churn'])
    axes[i].set_title(f'Count plot of {col}')  # Add title for each subplot
    axes[i].set_xlabel('')  # Remove x-axis label for better visual clarity
    axes[i].set_ylabel('Count')  # Keep the y-axis label

# Adjust layout to prevent overlapping
plt.tight_layout()

# Show the subplots
plt.show()
```

#customers who lack services such as Phone Service, Online Security, Device Protection, and Tech Support are more likely to churn, as indicated by the higher proportion of orange bars. Conversely, customers with these services tend to stay with the company. Additionally, churn rates are higher among Fiber Optic internet users and those without StreamingTV or StreamingMovies subscriptions.
